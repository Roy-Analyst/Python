# Data cleaning

```python
data = data.drop(['LoanNr_ChkDgt'], axis=1)
```

```python
data
```

```python
data = data.drop(['Name'], axis=1)
data = data.drop(['Bank'], axis=1)
data = data.drop(['Zip'], axis=1)
```

```python
data
```

```python
data = data.drop(['ApprovalDate'], axis=1)
data = data.drop(['DisbursementDate'], axis=1)
data = data.drop(['ChgOffDate'], axis=1)
data = data.drop(['City'], axis=1)
```

```python
data
```

```python
# cleaning the null values in MIS column
data = data.dropna(subset=['MIS_Status'])
```

```python
data
```

```python
data = data.dropna()
```

```python
data
```

```python
data.isnull().sum()
```

```python
# from the given data we can see that disbursementDate, DisbrusementGross, BalanceGross , chgoffPrinGr ,countinue...
# GrAppv, SBA_Appv has a symbol '$' in it. so removing special character
```

```python
ColumnsToBeModified = ['DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv']
```

```python
for col in ColumnsToBeModified:
    data[col] = data[col].str.replace('$', '').str.replace(',', '')
```

```python
data
```

```python
# From NAIC's Column defining the sector number seperately
data['NAICS'] = data['NAICS'].astype(str).apply(lambda x: x[:2])
```

```python
data.columns = data.columns.str.replace("NAICS", "Sector")
```

```python
data['Sector'].unique()
```

```python
# since MIS_Status column contains the target variable we are converting it into 0's and 1's
data['MIS_Status'] = data['MIS_Status'].apply(lambda x: 1 if x == 'P I F' else 0)
data.columns = data.columns.str.replace("MIS_Status", "TargetVariable")
```

```python
data['NewExist'].unique()
```

```python
## 0 = Existing business, 1 = New business
data['NewExist'] = data['NewExist'].astype(int).replace([1, 2], [0, 1])
```

```python
data.columns = data.columns.str.replace("NewExist", "Business Level")
```

```python
## Franchise code, (00000 or 00001) = No franchise

# FranciseCode is used to find wheather the company or indiviual has a francise or not  and from the given data we can
# say if its >1 the they have a francise else they dont have one. so we can represent it by 0 and 1
data['FranchiseCode'].unique()
```

```python
data['FranchiseCode']=data['FranchiseCode'].replace([(data['FranchiseCode'] <= 1), (data['FranchiseCode'] > 1)], [0, 1])
```

```python
data['FranchiseCode'].unique()
```

```python
data['FranchiseCode'] = data['FranchiseCode'].apply(lambda x: 1 if x > 1 else 0)
```

```python
data['FranchiseCode'].unique()
```

```python
# lowdoc is used to find the wheather a loan requires  minimal documentation for approval if yes the 1 else 0
# revlinecr is used to find wheater the loan is around the required credit if yes then 1 else 0

data['RevLineCr'].value_counts()
```

```python
data['LowDoc'].value_counts()
```

```python
data = data[(data.RevLineCr == 'Y') | (data.RevLineCr == 'N')]
data = data[(data.LowDoc == 'Y') | (data.LowDoc == 'N')]
```

```python
data['RevLineCr'].unique()
```

```python
data['RevLineCr'] = np.where(data['RevLineCr'] == 'N', 0, 1)
data['LowDoc'] = np.where(data['LowDoc'] == 'N', 0, 1)
```

```python
data['RevLineCr'].unique()
```

```python
data['ApprovalFY'].unique()
```

```python
# from the given data i found that there is an unwanted 'A' so formating into numeric
data['ApprovalFY'] = data['ApprovalFY'].apply(lambda x: int(str(x).replace('A', '')))
```

```python
data['ApprovalFY'].unique()
```

```python
# since the dataset has bank state and lenders state there might be some use in it so i am combining it
data.loc[(data["State"] == data['BankState']), 'SameStateOrNot'] = 0
data.loc[(data["State"] != data['BankState']), 'SameStateOrNot'] = 1
```

## Data Modification

```python
# the DisbursementGross and the gross approved can be compared because it says wheathar the borrower has
# borrowed all the aprooved money or less then that this will gravelly affect the prediction so i am combining it
data['BorrowedFullyOrNot'] = np.where(data['DisbursementGross'] == data['GrAppv'], 1, 0)
```

```python
data['SameStateOrNot'].dtypes
```

```python
# To make things easy for model to train, converting columns into integers and float
data['SameStateOrNot'] = data['SameStateOrNot'].astype(int)
```

```python

data['DisbursementGross'] = data['DisbursementGross'].astype(float)
data['GrAppv'] = data['GrAppv'].astype(float)
```

```python
data.drop(columns=['BalanceGross', 'SBA_Appv', 'ChgOffPrinGr'], inplace=True)
```

```python
data = pd.get_dummies(data)
```

```python
input_predictors = data.drop(['TargetVariable'], axis=1)
ouptut_target = data["TargetVariable"]
```

# Importing Module Training Libraries

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier
import time
import joblib
```

```python
x_train, x_val, y_train, y_val  = train_test_split(input_predictors, ouptut_target,
                                                    test_size = 0.25, random_state = 7)
```

```python
# x_val = x_train , y_val = y_test
```

```python
# MODEL-1) LogisticRegression
LogisticRegression = LogisticRegression()
start_time = time.time()
LogisticRegression.fit(x_train, y_train)
end_time = time.time()
y_pred = LogisticRegression.predict(x_val)
acc_LogisticRegression = round(accuracy_score(y_pred, y_val) * 100, 2)
```

```python
# Print accuracy and training time
print("\n\nMODEL-1: Accuracy of LogisticRegression: ", acc_LogisticRegression)
print("Training Time: {:.2f} seconds".format(end_time - start_time))
# MODEL-1: Accuracy of LogisticRegression:  82.07
# Training Time: 14.24 seconds

# Save the trained model to a file
model_filename = 'LogisticRegression_model.pkl'
joblib.dump(LogisticRegression, model_filename)

print(f"Trained model saved as {model_filename}")
```

```python
acc_LogisticRegression
```

```python
#MODEL-2) LinearDiscriminantAnalysis
LinearDiscriminantAnalysis = LinearDiscriminantAnalysis()
start_time = time.time()
LinearDiscriminantAnalysis.fit(x_train, y_train)
end_time = time.time()
y_pred = LinearDiscriminantAnalysis.predict(x_val)
acc_LinearDiscriminantAnalysis = round(accuracy_score(y_pred, y_val) * 100, 2)

# Print accuracy and training time
print("\n\nMODEL-5: Accuracy of LinearDiscriminantAnalysis: ", acc_LinearDiscriminantAnalysis)
print("Training Time: {:.2f} seconds".format(end_time - start_time))
# MODEL-5: Accuracy of LinearDiscriminantAnalysis:  83.88
# Training Time: 10.12 seconds

model_filename = 'LinearDiscriminantAnalysis_model.pkl'
joblib.dump(LinearDiscriminantAnalysis, model_filename)

print(f"Trained model saved as {model_filename}")
```

```python
#MODEL-3) Gaussian Naive Bayes

GaussianNB = GaussianNB()
start_time = time.time()
GaussianNB.fit(x_train, y_train)
end_time = time.time()
y_pred = GaussianNB.predict(x_val)
acc_GaussianNB = round(accuracy_score(y_pred, y_val) * 100, 2)
```
